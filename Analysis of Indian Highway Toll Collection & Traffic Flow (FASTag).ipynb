```python
# Assignment 1  Analysis of Indian Highway Toll Collection & Traffic Flow (FASTag)
# Google Colab Notebook
# Author: Shiv Mishra
# Date: Aug 16, 2025


#  Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt


#  Step 2: Create the dataset (from IHMCL ETC Transaction Reports, FY-wise aggregates)
data = [
    {"FY":"2016-17","Transactions_lakh":236,"Collection_crore":661},
    {"FY":"2017-18","Transactions_lakh":1271,"Collection_crore":3352},
    {"FY":"2018-19","Transactions_lakh":2544,"Collection_crore":5760},
    {"FY":"2019-20","Transactions_lakh":5830,"Collection_crore":11306},
    {"FY":"2020-21","Transactions_lakh":13269,"Collection_crore":22755},
    {"FY":"2021-22","Transactions_lakh":24352,"Collection_crore":38069},
    {"FY":"2022-23","Transactions_lakh":33887,"Collection_crore":54068},
    {"FY":"2023-24","Transactions_lakh":38152,"Collection_crore":64594},
    {"FY":"2024-25","Transactions_lakh":41638,"Collection_crore":72390},
    {"FY":"2025-26* (till Jul)","Transactions_lakh":15244,"Collection_crore":27134},
]
df = pd.DataFrame(data)

# Compute Avg Toll per Vehicle () = (Collection_crore / Transactions_lakh) * 100
df["Avg_Rs_per_vehicle"] = (df["Collection_crore"] / df["Transactions_lakh"]) * 100

# Show dataset
df

print("Analysis of Indian Highway Toll Collection & Traffic Flow (FASTag)")
print("Project prepared by: SHIV MISHRA\n")
#  Step 3: Visualizations

# Transactions trend
plt.figure(figsize=(8,5))
plt.plot(df["FY"], df["Transactions_lakh"], marker='o')
plt.title("Annual FASTag Transactions (lakh)")
plt.xlabel("Financial Year")
plt.ylabel("Transactions (lakh)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Collection trend
plt.figure(figsize=(8,5))
plt.plot(df["FY"], df["Collection_crore"], marker='o', color="green")
plt.title("Annual FASTag Toll Collection ( crore)")
plt.xlabel("Financial Year")
plt.ylabel("Collection ( crore)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Average toll per vehicle  last 3 FYs
last3 = df.iloc[6:9]  # FY2022-23 to FY2024-25
plt.figure(figsize=(6,4))
plt.bar(last3["FY"], last3["Avg_Rs_per_vehicle"], color="orange")
plt.title("Average Toll Collected per Vehicle ()")
plt.xlabel("Financial Year")
plt.ylabel(" per vehicle")
plt.show()

# ---
#  Step 4: Insights (print summary)


print("=== Key Insights (FY2022-23 to FY2024-25) ===")
print(f"1) Collections grew from {df.iloc[6]['Collection_crore']} cr to {df.iloc[8]['Collection_crore']} cr (+19%).")
print(f"2) Transactions grew from {df.iloc[6]['Transactions_lakh']} lakh to {df.iloc[8]['Transactions_lakh']} lakh (+23%).")
print(f"3) Average /vehicle rose from ~{df.iloc[6]['Avg_Rs_per_vehicle']:.0f} to ~{df.iloc[8]['Avg_Rs_per_vehicle']:.0f}.")
print(f"4) FY2025-26 (till Jul) already shows {df.iloc[9]['Collection_crore']} cr collections, suggesting another strong year.")
```

```python
# 2x2 subplot layout for all charts in one figure
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1) Transactions trend
axes[0,0].plot(df["FY"], df["Transactions_lakh"], marker='o')
axes[0,0].set_title("Annual FASTag Transactions (lakh)")
axes[0,0].set_xlabel("FY")
axes[0,0].set_ylabel("Transactions (lakh)")
axes[0,0].tick_params(axis='x', rotation=45)

# 2) Collections trend
axes[0,1].plot(df["FY"], df["Collection_crore"], marker='o', color="green")
axes[0,1].set_title("Annual Toll Collection ( crore)")
axes[0,1].set_xlabel("FY")
axes[0,1].set_ylabel(" crore")
axes[0,1].tick_params(axis='x', rotation=45)

# 3) Average toll per vehicle (all years)
axes[1,0].bar(df["FY"], df["Avg_Rs_per_vehicle"], color="orange")
axes[1,0].set_title("Average Toll per Vehicle ()")
axes[1,0].set_xlabel("FY")
axes[1,0].set_ylabel(" / vehicle")
axes[1,0].tick_params(axis='x', rotation=45)

# 4) Highlight last 3 years (zoomed bar chart)
last3 = df.iloc[6:9]
axes[1,1].bar(last3["FY"], last3["Avg_Rs_per_vehicle"], color="purple")
axes[1,1].set_title("Avg Toll per Vehicle  Last 3 FYs")
axes[1,1].set_xlabel("FY")
axes[1,1].set_ylabel(" / vehicle")
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
