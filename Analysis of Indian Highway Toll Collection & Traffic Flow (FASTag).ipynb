```python
# Assignment 1 â€“ Analysis of Indian Highway Toll Collection & Traffic Flow (FASTag)
# Google Colab Notebook
# Author: Shiv Mishra
# Date: Aug 16, 2025


# ðŸ”¹ Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt


# ðŸ”¹ Step 2: Create the dataset (from IHMCL ETC Transaction Reports, FY-wise aggregates)
data = [
    {"FY":"2016-17","Transactions_lakh":236,"Collection_crore":661},
    {"FY":"2017-18","Transactions_lakh":1271,"Collection_crore":3352},
    {"FY":"2018-19","Transactions_lakh":2544,"Collection_crore":5760},
    {"FY":"2019-20","Transactions_lakh":5830,"Collection_crore":11306},
    {"FY":"2020-21","Transactions_lakh":13269,"Collection_crore":22755},
    {"FY":"2021-22","Transactions_lakh":24352,"Collection_crore":38069},
    {"FY":"2022-23","Transactions_lakh":33887,"Collection_crore":54068},
    {"FY":"2023-24","Transactions_lakh":38152,"Collection_crore":64594},
    {"FY":"2024-25","Transactions_lakh":41638,"Collection_crore":72390},
    {"FY":"2025-26* (till Jul)","Transactions_lakh":15244,"Collection_crore":27134},
]
df = pd.DataFrame(data)

# Compute Avg Toll per Vehicle (â‚¹) = (Collection_crore / Transactions_lakh) * 100
df["Avg_Rs_per_vehicle"] = (df["Collection_crore"] / df["Transactions_lakh"]) * 100

# Show dataset
df

print("Analysis of Indian Highway Toll Collection & Traffic Flow (FASTag)")
print("Project prepared by: SHIV MISHRA\n")
# ðŸ”¹ Step 3: Visualizations

# Transactions trend
plt.figure(figsize=(8,5))
plt.plot(df["FY"], df["Transactions_lakh"], marker='o')
plt.title("Annual FASTag Transactions (lakh)")
plt.xlabel("Financial Year")
plt.ylabel("Transactions (lakh)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Collection trend
plt.figure(figsize=(8,5))
plt.plot(df["FY"], df["Collection_crore"], marker='o', color="green")
plt.title("Annual FASTag Toll Collection (â‚¹ crore)")
plt.xlabel("Financial Year")
plt.ylabel("Collection (â‚¹ crore)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Average toll per vehicle â€“ last 3 FYs
last3 = df.iloc[6:9]  # FY2022-23 to FY2024-25
plt.figure(figsize=(6,4))
plt.bar(last3["FY"], last3["Avg_Rs_per_vehicle"], color="orange")
plt.title("Average Toll Collected per Vehicle (â‚¹)")
plt.xlabel("Financial Year")
plt.ylabel("â‚¹ per vehicle")
plt.show()

# ---
# ðŸ”¹ Step 4: Insights (print summary)


print("=== Key Insights (FY2022-23 to FY2024-25) ===")
print(f"1) Collections grew from â‚¹{df.iloc[6]['Collection_crore']} cr to â‚¹{df.iloc[8]['Collection_crore']} cr (+19%).")
print(f"2) Transactions grew from {df.iloc[6]['Transactions_lakh']} lakh to {df.iloc[8]['Transactions_lakh']} lakh (+23%).")
print(f"3) Average â‚¹/vehicle rose from ~â‚¹{df.iloc[6]['Avg_Rs_per_vehicle']:.0f} to ~â‚¹{df.iloc[8]['Avg_Rs_per_vehicle']:.0f}.")
print(f"4) FY2025-26 (till Jul) already shows â‚¹{df.iloc[9]['Collection_crore']} cr collections, suggesting another strong year.")
```
